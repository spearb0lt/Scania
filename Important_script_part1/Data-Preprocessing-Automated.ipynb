```python
import pandas as pd

# File paths
# csv1_path = r"C:\Users\ASUS\Desktop\SCANIA\2024-34-2\2024-34-2\data\train_operational_readouts.csv"
# csv2_path = r"C:\Users\ASUS\Desktop\SCANIA\2024-34-2\2024-34-2\data\train_tte.csv"

# Read CSV files
# df1 = pd.read_csv(csv1_path)
df1=df
df2 = pd.read_csv(csv2_path)

# Add class_label column initialized with -1 (assuming -1 is not a valid class_label)
df1['in_study_repair'] = pd.NA
df1['in_study_repair'] = df1['in_study_repair'].astype('Int64')  
# Ensure class_label in df2 is int64
df2['in_study_repair'] = df2['in_study_repair'].astype('int64')

# Iterate over unique vehicle_ids
for vehicle_id in df1['vehicle_id'].unique():
    # Get the last occurrence index of the vehicle_id in df1
    last_index = df1[df1['vehicle_id'] == vehicle_id].index[-1]
    
    # Get the corresponding class_label from df2
    in_study_repair = df2[df2['vehicle_id'] == vehicle_id]['in_study_repair'].values
    
    if len(in_study_repair) >= 0:  # Ensure class_label exists for the vehicle_id in df2
        df1.at[last_index, 'in_study_repair'] = in_study_repair[0]

# Convert class_label back to int64 if needed
# df1['in_study_repair'] = df1['in_study_repair'].astype('int64')

# Save the updated CSV
# Save the updated CSV
# df1.to_csv(r'C:\Users\ASUS\Desktop\SCANIA\2024-34-2\2024-34-2\data\updated_train_operational_readouts.csv', index=False)
```

```python
df=df1
# Create row_number column
df['row_number'] = range(1, len(df) + 1)

# Create class_label column with Int64 type and NaN values
df['class_label'] = pd.NA

# Identify the last row of each vehicle_id
df['is_last_row'] = df['vehicle_id'].ne(df['vehicle_id'].shift(-1))

# Apply conditions for class_label
df.loc[df['is_last_row'] & (df['in_study_repair'] == 0), 'class_label'] = 0
df.loc[df['is_last_row'] & (df['in_study_repair'] == 1), 'class_label'] = 4

# Drop the helper column
# df.drop(columns=['is_last_row'], inplace=True)

# Save the updated DataFrame (optional)
# df.to_csv(csv_path, index=False)
```

```python
def assign_class_labels(group):
    last_row_index = group.index[-1]
    class_label_last = group.loc[last_row_index, 'class_label']
    
    if class_label_last == 0:
        group['class_label'] = 0

    elif class_label_last == 1:
        group['class_label'] = 0
        group.iloc[-23:, group.columns.get_loc('class_label')] = 1

    elif class_label_last == 2:
        group['class_label'] = 0
        group.iloc[-35:-11, group.columns.get_loc('class_label')] = 1  # next 24
        group.iloc[-11:, group.columns.get_loc('class_label')] = 2     # last 11

    elif class_label_last == 3:
        group['class_label'] = 0
        group.iloc[-41:-17, group.columns.get_loc('class_label')] = 1  # next 24
        group.iloc[-17:-5, group.columns.get_loc('class_label')] = 2   # next 12
        group.iloc[-5:, group.columns.get_loc('class_label')] = 3      # last 5

    elif class_label_last == 4:
        group['class_label'] = 0
        group.iloc[-48:-24, group.columns.get_loc('class_label')] = 1  # next 24
        group.iloc[-24:-12, group.columns.get_loc('class_label')] = 2  # next 12
        group.iloc[-12:-6, group.columns.get_loc('class_label')] = 3   # next 6
        group.iloc[-6:, group.columns.get_loc('class_label')] = 4      # last 5

    return group


# Apply function to each vehicle_id
df = df.groupby('vehicle_id', group_keys=False).apply(assign_class_labels)

# Drop the helper column
df.drop(columns=['is_last_row'], inplace=True)

# Save the updated DataFrame (optional)
# df.to_csv(csv_path, index=False)
```
