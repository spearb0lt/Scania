min max scaling numeric counters

scaler = MinMaxScaler()  # Initialize once before the loop
for i in column_list[2:]:
    df[i] = scaler.fit_transform(df[[i]])  # Apply the same scaler instance to all columns
 Effect:

A single instance of MinMaxScaler is used for all columns.
The scaling is applied globally, meaning all columns are transformed using the same min and max values.

for i in column_list[2:]:
    scaler = MinMaxScaler()  # A new scaler instance is created for each column
    df[i] = scaler.fit_transform(df[[i]])  # Each column is scaled independently

 Effect:

A new instance of MinMaxScaler is created for each column.
Each column is scaled independently, meaning fit_transform() computes min-max scaling using only that column's values.


Key Differences:
Aspect	First Code	Second Code
Scalability	Uses one scaler instance for all columns	Creates a new instance for each column
Consistency	Min & Max computed globally across all columns	Min & Max computed separately for each column
Memory Efficiency	More efficient (single object)	Less efficient (multiple objects)
Result	All columns scale using the same range of min-max values	Each column scales independently


Which One to Use?
 Use the first approach if you want to scale all columns consistently using the same global min-max.
 Use the second approach if you want each column to be scaled independently, meaning each will have its own 0-1 range.

```python
import pandas as pd

# Load the CSV files
csv1_path = r"C:\Users\ASUS\Desktop\SCANIA\2024-34-2\2024-34-2\data\validation_operational_readouts.csv"
csv2_path = r"C:\Users\ASUS\Desktop\SCANIA\2024-34-2\2024-34-2\data\validation_labels.csv"

df1 = pd.read_csv(csv1_path)  # Operational readouts
df2 = pd.read_csv(csv2_path)  # Labels

# Merge df2 into df1 on 'vehicle_id' to get class_label
df1 = df1.merge(df2[['vehicle_id', 'class_label']], on='vehicle_id', how='left')

# Function to assign class_label based on last row's value
def assign_labels(group):
    last_idx = group.index[-1]  # Get the last row index
    last_value = group.loc[last_idx, 'class_label']  # Get last row class_label from csv2
    
    # Default all values to NaN
    group['class_label'] = float('nan')

    # Assign the last row its class_label from csv2
    group.loc[last_idx, 'class_label'] = last_value

    # Assign values based on the last row's class_label
    if last_value == 4:
        group.loc[max(last_idx - 6, group.index[0]):last_idx - 1, 'class_label'] = 3
        group.loc[max(last_idx - 18, group.index[0]):last_idx - 7, 'class_label'] = 2
        group.loc[max(last_idx - 42, group.index[0]):last_idx - 19, 'class_label'] = 1
        group.loc[:max(last_idx - 43, group.index[0] - 1), 'class_label'] = 0
    elif last_value == 3:
        group.loc[max(last_idx - 12, group.index[0]):last_idx - 1, 'class_label'] = 2
        group.loc[max(last_idx - 36, group.index[0]):last_idx - 13, 'class_label'] = 1
        group.loc[:max(last_idx - 37, group.index[0] - 1), 'class_label'] = 0
    elif last_value == 2:
        group.loc[max(last_idx - 24, group.index[0]):last_idx - 1, 'class_label'] = 1
        group.loc[:max(last_idx - 25, group.index[0] - 1), 'class_label'] = 0
    elif last_value == 1:
        group.loc[:last_idx - 1, 'class_label'] = 0
    elif last_value == 0:
        group['class_label'] = 0  # Everything remains 0

    return group

# Apply the function to each vehicle_id group
df1 = df1.groupby("vehicle_id", group_keys=False).apply(assign_labels)

# Save the updated DataFrame
df1.to_csv(csv1_path, index=False)

print("Updated CSV1 with propagated class_label based on last row value from CSV2.")
```

```python
def assign_class_labels(group):
    last_row_index = group.index[-1]
    class_label_last = group.loc[last_row_index, 'class_label']
    
    if class_label_last == 0:
        group['class_label'] = 0
    elif class_label_last == 1:
        group.loc[last_row_index - 48:last_row_index - 24, 'class_label'] = 1
        group['class_label'].fillna(0, inplace=True)
    elif class_label_last == 2:
        group.loc[last_row_index - 48:last_row_index - 24, 'class_label'] = 1
        group.loc[last_row_index - 24:last_row_index - 12, 'class_label'] = 2
        group['class_label'].fillna(0, inplace=True)
    elif class_label_last == 3:
        group.loc[last_row_index - 48:last_row_index - 24, 'class_label'] = 1
        group.loc[last_row_index - 24:last_row_index - 12, 'class_label'] = 2
        group.loc[last_row_index - 12:last_row_index - 6, 'class_label'] = 3
        group['class_label'].fillna(0, inplace=True)
    elif class_label_last == 4:
        group.loc[last_row_index - 48:last_row_index - 24, 'class_label'] = 1
        group.loc[last_row_index - 24:last_row_index - 12, 'class_label'] = 2
        group.loc[last_row_index - 12:last_row_index - 6, 'class_label'] = 3
        group.loc[last_row_index - 6:last_row_index, 'class_label'] = 4
        group['class_label'].fillna(0, inplace=True)
    return group

# Apply function to each vehicle_id
df = df.groupby('vehicle_id', group_keys=False).apply(assign_class_labels)

# Drop the helper column
df.drop(columns=['is_last_row'], inplace=True)

# Save the updated DataFrame (optional)
# df.to_csv(csv_path, index=False)
```
